```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
ipl_df= pd.read_csv("IPL Matches 2008-2020 (1).csv")
```

```python
#HEAT MAP
sn.heatmap(ipl_df.isnull(),cmap='Blues')
```

```python
#COUNT PLOT

pom = ipl_df['player_of_match'].value_counts().iloc[:10]
plt.figure(figsize=(4,4))
c1 = sn.countplot(y= 'player_of_match',order=pom.index,data = ipl_df)
c1.bar_label(c1.containers[0],size = 7)
plt.ylabel('Player',fontsize=8)
plt.xlabel('No: of matches',fontsize=8)
plt.title('Top 10 Players who received POM',fontsize=10)
plt.show()
```

```python
#AB de Villiers has received 23 player of the match awards over his entire IPL playing career.
```

```python
#pie chart
labels = ipl_df['toss_decision'].unique()
sizes =  ipl_df['toss_decision'].value_counts()
colors = ['grey','orange']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Percentage of the Teams won the toss and their decision",fontsize=10)
plt.show()
```

```python
#BAR PLOT
res = ipl_df['result'].value_counts()
plt.figure(figsize=(4,4))
c1 = sn.countplot(x='result',order=res.index,data = ipl_df)
c1.bar_label(c1.containers[0],size = 8)
plt.ylabel('No of matches won',fontsize=8)
plt.xlabel('Results',fontsize=8)
plt.title('Results of the matches played',fontsize=10)
plt.show()

sn.distplot(ipl_df['id'])

sn.boxplot(ipl_df['id'])
```
